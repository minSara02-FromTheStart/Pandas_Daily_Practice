```python
#What does our dataset look like?
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("NetflixData.csv")

#print("(rows,columns): ", df.shape)
#print("Columns : ", df.columns)
#print("Data Types: ", type(df))

print("Data Info: ", df.info())
```

```python
#Is Netflix content mostly Movies or TV Shows?
type_count = df["type"].value_counts()
print("What type of content dominates the platform?")
print("Ans : ")
print(type_count)
```

```python
plt.figure(figsize=(3,2))
type_count.plot(kind="bar")
plt.title("Movie Vs Tv Shows")
plt.xlabel("Content Type")
plt.ylabel("Content Number")

plt.show()
```

```python
plt.figure()
plt.title("Movie,TV Show & Documentary Distribution")
type_count.plot(kind="pie",autopct="%1.1f%%", startangle=90,figsize=(3,2))
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
#Which genre appears the most?
genre_count= df["genre"].value_counts()
print(" Genre: ", genre_count)
```

```python
#Visiulization

plt.figure(figsize=(3,2))
genre_count.plot(kind="bar")
plt.title("Genre Visualization ")
plt.xlabel("Genre Type")
plt.ylabel("Genre Count")
plt.show()
```

```python
plt.figure()
genre_count.plot(kind="pie", autopct = "%1.1f%%", startangle = 90, figsize = (7,7),colors=plt.cm.tab20.colors)
plt.title("Genre Visualization ")
plt.ylabel("")
plt.show()
```

```python
#What ratings are most common?
rating_count = df["rating"].value_counts()
print("Rating : ", rating_count)
```

```python
#Visiulization

plt.figure(figsize=(3,2))
rating_count.plot(kind="bar")
plt.title("Rating Visualization ")
plt.xlabel("Rating Type")
plt.ylabel("Rating Count")
plt.show()
```

```python
#Is content increasing over time?
year_counts = df["release_year"].value_counts().sort_index()
print(year_counts)
```

```python
plt.figure(figsize=(3,2))
year_counts.plot(kind="line")
plt.title("Number of Titles Released Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
#Which countries produce the highest number of titles?  
country_counts = df["country"].value_counts()
top5_countries = country_counts.head(5)
print(top5_countries)
```

```python
plt.figure(figsize = (3,2))
top5_countries.plot(kind="bar")
plt.title("Top 5 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.show()
```
