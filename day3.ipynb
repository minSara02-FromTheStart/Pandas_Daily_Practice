```python
categorical_cols = ["gender"]

for c in categorical_cols:
    if c in df.columns:
        df[c] = df[c].astype("category")
```

```python
gender_counts = df["gender"].value_counts(dropna=True)
print(gender_counts)
```

```python
gender_counts_sorted_by_value = gender_counts.sort_values(ascending=True)
print(gender_counts_sorted_by_value)
```

```python
gender_counts_sorted_by_index = gender_counts.sort_index()
print(gender_counts_sorted_by_index)
```

```python
import matplotlib.pyplot as plt

plt.figure()
gender_counts.sort_values(ascending=False).plot(kind="bar")
plt.title("Number of Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_counts = df["gender"].value_counts()

plt.figure()
gender_counts.plot(kind="pie")
plt.title("Student Distribution by Gender")
plt.ylabel("")   
plt.show()
```

```python
plt.figure()
gender_counts.plot(
    kind="pie",
    autopct="%1.1f%%",   
    startangle=90        
)
plt.title("Student Distribution by Gender")
plt.ylabel("")
plt.show()
```
