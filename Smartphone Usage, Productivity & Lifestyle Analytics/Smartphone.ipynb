```python
import pandas as pd


file_path = r"C:\Users\tsnmi\GitHub_Work\pandas-daily-practice\Smartphone Usage, Productivity & Lifestyle Analytics\archive\Smartphone_Usage_Productivity_Dataset_50000.csv"

df = pd.read_csv(file_path)  #Load a CSV file
```

```python
df.isnull().sum()     #Find how many missing values exist in each column.
```

```python
#Remove all rows where any value is missing.

df.dropna(inplace=True)   #in my this data there is no missing value that's why the shape is same as before
df.shape
```

```python
#Replace missing values in the Age column with the mean.

df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
#Replace missing values in the Occupation column with "Unknown"
df['Occupation'] = df['Occupation'].fillna("Unknown")
```

```python
#Drop a column named Daily_Phone_Hours
df.drop('Daily_Phone_Hours',axis=1,inplace = True)
```

```python
#Count total missing values in the entire dataset.

df.isnull().sum().sum()
```
