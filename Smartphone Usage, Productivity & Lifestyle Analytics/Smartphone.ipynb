```python
import pandas as pd

file_path = Path.cwd() /"archive" / "Smartphone_Usage_Productivity_Dataset_50000.csv"

df = pd.read_csv(file_path)  #Load a CSV file
```

```python
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 120)
df.head(10)  #Display the first 10 rows of the dataset.
```

```python
df.shape  #Check the shape (rows, columns) of the dataset.
```

```python
df.dtypes     #Show the data types of all columns.
```

```python
df.isnull().sum()     #Find how many missing values exist in each column.
```

```python
#Remove all rows where any value is missing.

df.dropna(inplace=True)   #in my this data there is no missing value that's why the shape is same as before
df.shape
```

```python
#Replace missing values in the Age column with the mean.

df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
#Replace missing values in the Occupation column with "Unknown"
df['Occupation'] = df['Occupation'].fillna("Unknown")
```

```python
#Drop a column named Daily_Phone_Hours
df.drop('Daily_Phone_Hours',axis=1,inplace = True)
```

```python
#Count total missing values in the entire dataset.

df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#Standardize a Gender column that contains values like:
# "Male", "male", "M", "m"
# "Female", "female", "F", "f"
# Make them consistent.

df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] =df['Gender'].replace({'male':'Male','m':'Male','female':'Female','f':'Female'})
```

```python
#Convert all column names to lowercase.

df.columns = df.columns.str.lower()

df.dtypes
```

```python
df['Daily_Phone_Hours'] = pd.to_numeric(df['Daily_Phone_Hours'],errors='coerce')   #string-->numeric
```
