```python
import pandas as pd


data = {
    "Name": ["Asha", "Rahim", "Sara"],
    "Age": [20, 22, 21],
    "Score": [85, 90, 88]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["passed"] = [True,True,True]

print(df)
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
print("Data types:\n", df.dtypes)
```

```python
#change one name in the "Name" column

df.loc[2,"Name"] = "Maliat"
print(df)
```

```python
#print only the "Name and Score" colums

print(df[["Name", "Score"]])
```

```python
#add one more student row
df.loc[len(df)] = ["Nila", 23 , 92, True]
print(df)
```
