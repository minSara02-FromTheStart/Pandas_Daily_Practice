```python
class Student:
    school_name = "Future Academy"
    
    def __init__(self, name, cgpa):
        self.name = name
        self.cgpa = cgpa
    
    def display_info(self):
        print(f"Name: {self.name}, CGPA: {self.cgpa}, School: {self.school_name}")
    
    def update_cgpa(self, new_cgpa):
        self.cgpa = new_cgpa
    
    def honor_roll(self):
        return self.cgpa >= 3.7
    
s1 = Student("Alice", 3.8)
s2 = Student("Bob", 3.5)
s2.update_cgpa(3.8)
print(s2.honor_roll()) 
```
